```python
# Calculate cosine similarity manually
def cosine_similarity_manual(vec1, vec2):
    dot_product = sum(p*q for p,q in zip(vec1, vec2))
    magnitude1 = sum(p**2 for p in vec1) ** .5
    magnitude2 = sum(q**2 for q in vec2) ** .5
    if magnitude1 == 0 or magnitude2 == 0:
        return 0
    else:
        return dot_product / (magnitude1 * magnitude2)

# Calculate cosine similarity manually
cosine_sim_manual = cosine_similarity_manual(vector_g, vector_f_l)
print(vector_g)
print(vector_f_l)
print(cosine_sim_manual)
```

```python
import numpy as np
import random
import re
from collections import Counter
!pip install jieba
import jieba

def extract_category(filename):
    return filename.split('_')[-1] #.split('.')[0]



class Bigrameda:
    def __init__(self):
        self.bigramcounts = Counter()
        self.bigramprob = {}
        self.wordList = []

    def bigramCount(self, cleaneddata):
        jieba.load_userdict('mydictionary.txt')
        word = jieba.cut(cleaneddata, cut_all=False)
        self.wordList = [w for w in word if len(w) >= 2]
        self.bigramcounts = Counter(self.wordList)
        return sum(self.bigramcounts.values())        

    def ttr(self, sample_size=None, simulations=None):
        #all_bigrams = list(self.bigramcounts) #.elements())
        #print(len(all_bigrams))
        print(len(self.wordList))#, self.wordList)
        ttr_samples = []
        for i in range(simulations):
            if len(self.wordList) < sample_size:
                sample = self.wordList  # 如果文本太短，使用全部文本
            else:
                sample = random.choices(self.wordList, k=sample_size)
                #print(sample)
            
            sample_types = len(set(sample))
            ttr_samples.append(sample_types)
        average = sum(ttr_samples) / len(ttr_samples)
        sd = np.std(ttr_samples, ddof=1)    
        return f"{average:.0f}, {sd:.0f}"
```

```python
class Unigrameda:
    def __init__(self):
        self.unigramcounts = Counter()
        self.unigramprob = {}
        self.uniwordList = []

    def unigramCount(self, cleaneddata):
        #print(cleaneddata)
        self.uniwordList = list(cleaneddata)
        #print(self.uniwordList)
        self.unigramcounts = Counter(cleaneddata)
        return sum(self.unigramcounts.values())

    def ttr(self, sample_size=None, simulations=None):
        #print(len(self.uniwordList))#, self.uniwordList)
        ttr_samples = []
        for i in range(simulations):
            if len(self.uniwordList) < sample_size:
                sample = self.uniwordList  # 如果文本太短，使用全部文本
            else:
                sample = random.choices(self.uniwordList, k=sample_size)
            sample_types = len(set(sample))
            ttr_samples.append(sample_types)
        average = sum(ttr_samples) / len(ttr_samples)
        sd = np.std(ttr_samples, ddof=1)    
        return f"{average:.0f}, {sd:.0f}"
```

## TFIDF Example

```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [
'This is the first document.',
'This document is the second document.',
'And this is the third one.',
'Is this the first document?'
]
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)
feature_names = vectorizer.get_feature_names_out()
scores = X.toarray()

# 顯示關鍵字及其分數
for doc_idx, doc_scores in enumerate(scores):
    print(f"文檔 {doc_idx + 1}:")
    #print(doc_scores)
    for word_idx, score in enumerate(doc_scores):
        if score > 0:  # 為了簡潔起見，僅顯示有分數的單詞
            print(f"{feature_names[word_idx]}: {score}")
```
